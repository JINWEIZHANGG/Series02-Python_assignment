```python
# package imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters
frequency = 1         # frequency in Hz
amplitude = 1         # amplitude
phase = 0             # phase in radians
duration = 5          # duration in seconds also number of cycles
sampling_rate = 1000  # sampling rate in Hz
displacement = 0      # vertical shift
shift = 0.5           # lateral shift in seconds

# define function to compute sinusoid wave
def sinusoid_wave(frequency, amplitude, phase, duration, sampling_rate, displacement):
    t = np.arange(0, duration, 1/sampling_rate)
    y = amplitude * np.sin(2 * np.pi * frequency * t + phase) + displacement
    return t, y

# plotting function
def plot_wave(t, y, title):
    plt.figure(figsize=(10, 4))
    plt.plot(t, y)
    plt.title(title)
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    


sin_wave = sinusoid_wave(frequency, amplitude, phase, duration, sampling_rate, displacement)
plot_wave(sin_wave[0], sin_wave[1], 'Sinusoid Wave')
```
